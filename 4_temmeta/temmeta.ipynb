```python
from pyiron_experimental import Project
```

```python
pr = Project('tem')
```

```python
pr.remove_jobs(silently=True, recursive=True)
```

### Step 1: Create a TEMMETA job

```python
tem_job = pr.create_job(pr.job_type.TEMMETAJob, 'tem')
```

### Step 2: Load in the data file

```python
tem_job.file_name = "./experiment.emd"
```

Easily visualise the data

```python
tem_job.plot()
```

### Step 3: Select a vector

```python
tem_job.vector = [(35, 101), (205, 101)] 
```

```python
tem_job.plot()
```

```python
tem_job.input
```

Run the job

```python
tem_job.run()
```

### Step 4: Plot the profile along the selected vector

```python
tem_job.plot_profile()
```

```python
tem_job['output/generic/profile']
```
